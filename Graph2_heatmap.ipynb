```python
crime_final = ['Arson', 'Assault', 'Disorderly Conduct', 'Drug Offense',
      'Lost Property', 'Robbery', 'Sex Offense', 'Stolen Property',
      'Weapons Carrying Etc', 'Weapons Offense']
# neighborhood = ['Financial District/South Beach', 'Mission', 'Glen Park',
#       'Oceanview/Merced/Ingleside']
neighborhood = ['Financial District/South Beach', 'Mission', 'Castro/Upper Market',
                'Tenderloin', 'Glen Park', 'Oceanview/Merced/Ingleside']

# 'Glen Park',# 'Glen Park',
```

```python
df_final = df_crime_2019_jan_aug.loc[df_crime_2019_jan_aug['Incident Category'].isin(crime_final)]
df_final = df_final.loc[df_final['Analysis Neighborhood'].isin(neighborhood)]
```

```python
def fixDFcolsForR(myDF):
    for i in range(myDF.shape[1]): # for each col
        if myDF.iloc[:, i].dtype == 'O': # check if col is type object
            myDF.iloc[:, i] = myDF.iloc[:, i].astype(str) # if yes set type to string
    return myDF
df_final = fixDFcolsForR(df_final)
```

```python
# df_final.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
```

```python
df_final['Incident_category'] = df_final['Incident Category']
df_final['Neighborhood'] = df_final['Analysis Neighborhood']
```

```python
df_2 = df_final.groupby(['Neighborhood','Incident_category']).agg({'Incident Datetime':'count'}).reset_index()
df_2['values']=df_2['Incident Datetime']
df_2['Incident_category'].unique()
```

```python
%R -i df_2
```

```python
%%R -w 8 -h 8 --units in -r 400
ggplot(df_2, aes(x=Neighborhood,y= Incident_category)) +
geom_tile(aes(fill=values)) +
 scale_fill_gradient2(low="navy", mid="white", high="red")+
theme(legend.title = element_blank(),
     axis.title=element_blank(),
     panel.background = element_blank(),
     axis.ticks=element_blank(),
     axis.text.x = element_text( angle =90,size=7))
# scale_fill_gradient()
```
